```python
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
def image_to_ascii(image_path):
  image = plt.imread(image_path)
  plt.imshow(image, cmap='gray')
  plt.axis(False);
  if image.shape[2]==4:
    gray = cv.cvtColor(image, cv.COLOR_RGBA2GRAY)
  elif image.shape[2]==3:
    gray = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
  small_img = cv.pyrDown(cv.pyrDown(cv.pyrDown(gray)))
  string = ''
  for i in range(small_img.shape[0]):
    for j in range(small_img.shape[1]):
      if small_img[i,j]<0.25:
        string = string+('@@')
      elif small_img[i,j]<0.5:
        string = string+('//')
      elif small_img[i,j]<0.75:
        string = string+('::')
      else:
        string = string+('..')
    string = string+('\n')
  # print(string)
  file = open('output.txt',mode='w')
  file.write(string)
  file.close
  print(string)
```

```python
image_to_ascii('/content/drive/MyDrive/imports-for-colab/Photo_SRK.png')
```
